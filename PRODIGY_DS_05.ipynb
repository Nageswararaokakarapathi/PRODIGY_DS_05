```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("Accidents_2016.csv.zip")

df.head()
```

```python
df.info()

df.describe()

df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(data=df,
              x="Accident_Severity",
              palette="Set2")

plt.title("Accident Severity Distribution")

plt.savefig("images/accident_severity.png",dpi=300)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    y="Weather_Conditions",
    data=df,
    order=df["Weather_Conditions"].value_counts().index[:10]
)

plt.title("Top Weather Conditions")

plt.savefig("images/accidents_by_weather.png",dpi=300)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    y="Road_Surface_Conditions",
    data=df,
    order=df["Road_Surface_Conditions"].value_counts().index
)

plt.title("Road Surface Conditions")

plt.savefig("images/accidents_by_road_surface.png",dpi=300)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    y="Light_Conditions",
    data=df,
    order=df["Light_Conditions"].value_counts().index
)

plt.title("Light Conditions")

plt.savefig("images/accidents_by_light.png",dpi=300)

plt.show()
```

```python
numeric=df.select_dtypes(include=np.number)

plt.figure(figsize=(12,8))

sns.heatmap(
    numeric.corr(),
    cmap="coolwarm",
    annot=False
)

plt.title("Correlation Heatmap")

plt.savefig("images/correlation_heatmap.png",dpi=300)

plt.show()
```
